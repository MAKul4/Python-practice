**String & string opertions**

```python
# Assignment

my_sentence = "It's a wonderful day!"
```

```python
# Multiplication by a number will repeat the string. 
# Notice there is no space in between two strings

my_sentence * 2
```

```python
# Concatenate i.e. adding a word or sentence to a string

my_sentence + "Right ?"
```

```python
#Add a space
my_sentence + " " + "Right ?"
```

```python
# Find a letter or word in the strign

"a" in my_sentence
```

```python
"m" in my_sentence
```

```python
# Convert to lowercase

my_sentence.lower()
```
